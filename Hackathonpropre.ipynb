```python
import pandas as pd
import seaborn as sns
url = 'https://happiness-report.s3.amazonaws.com/2023/DataForTable2.1WHR2023.xls'
data = pd.read_excel(url)
data.rename(columns={'Country name':'Countryname'},inplace=True)
```

```python
dfItalie=data.query("Countryname=='Italy'")
dfCiv=data.query("Countryname=='Ivory Coast'")
dfBenin=data.query("Countryname=='Benin'")
dfCiv2=dfCiv.loc[944:]
dfItalie2=dfItalie.loc[933:]
dfBenin2=dfBenin.loc[182:]
data2=pd.concat([dfCiv2, dfItalie2,dfBenin2]) 
sns.set_style('darkgrid')
sns.relplot(data=data2,x='Log GDP per capita',y='Freedom to make life choices',hue='Countryname', style='Countryname')
```

Le but ici est de mettre en lumière une première fois le fait qu'un indicateur économique classique (ici le "GDP" donc le PIB) ne semble pas représenter réellement le niveau de bonheur des individus. On observe en effet ici que pour un PIB par tête plus élevé en Italie qu'au Bénin et en Côte d'Ivoire, la liberté de faire des choix de vie est plus faible pour autant.

```python
data4=pd.concat([dfItalie2,dfBenin2])
sns.displot(data=data4,x='Freedom to make life choices',hue='Countryname',kind='kde');
```

On remarque la même tendance en organisant les données diféremment.

```python
dfBot=data.query("Countryname=='Botswana'")
dfBot2=dfBot.loc[231:]
dfBot3=dfBot2.drop([235])
dfBot3.plot(x="Log GDP per capita", y="Social support", label="Social support", color="blue",legend=True, xlabel="log PIB par tête")
```

De la même manière, on remarque que dans un pays comme le Botswana l'aide sociale diminue globalement lorsque le PIB augmente, soulignant le fait que l'indicateur économique seul ne suffit pas à démontrer le niveau de vie et de bonheur de la population.
