Le but de ce notebook est de montrer que les donnés sur le bonheur doivent être prises en compte au même titre que les données économiques quand on analyse un pays. 
En particulier, on pense souvent qu'un pays va bien quand son PNB va bien. 
L'objectif ici est de mettre en évidence les liens existants ou non existants entre bonheur et PNB. 

```python
import pandas as pd
```

```python
url ='https://happiness-report.s3.amazonaws.com/2023/DataForTable2.1WHR2023.xls'
```

```python
import seaborn as sns 
```

```python
pip install xlrd
```

```python
data= pd.read_excel(url)
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
data.rename(columns={'Country name':'Countryname'},inplace=True)
```

```python
dfSingap=data.query("Countryname=='Singapore'")
```

```python
dfIr=data.query("Countryname=='Ireland'")
```

```python
dfSingap.head()
```

```python
dfIr.head()
```

```python
data2=pd.concat([dfSingap,dfIr])
data2
```

```python
sns.relplot(data=data2,x='year', y='Log GDP per capita', hue='Countryname')
```

```python
sns.relplot(data=data2,x='year', y='Positive affect', hue='Countryname')
```

Ces deux graphes montrent que pour 2 pays pour qui ont la même tendance de PNB au cours des années, les tendances de 'positive affect' sont différentes.
En effet l'Irelande et Singapoure ont des PNB similaires en ordre de grandeur et leur PNB croient au cours du temps. 
Cependant, les habitants de Singapoure sont en moyenne moins heureux que les habitants de l'Irelande. Ce qui montre qu'il n'y a pas corrélation exacte entre PNB et niveau de bonheur dans un pays.
